# **Step 4: Time Series Data**

Time series data is a collection of observations obtained through repeated measurements over time. 
Plot the points on a graph, and one of your axes would always be time.

# **Time series data can be useful for:**



*   Tracking daily, hourly, or weekly weather data
*   Tracking changes in application performance

*   Medical devices to visualize vitals in real time
*   Tracking network logs








```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates #for Date visualization
```

```python
plt.plot(MSFT.index, MSFT["Adj Close"])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
#Subplots
f, ax = plt.subplots(2, 2, figsize=(10,10), sharex=True)
f.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y"))   
f.gca().xaxis.set_major_locator(mdates.YearLocator())   

ax[0,0].plot(AAPL.index, AAPL["Adj Close"], color="r")
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title("AAPL")

ax[1,0].plot(GOOG.index, GOOG["Adj Close"], color="b")
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title("GOOG")

ax[0,1].plot(TSLA.index, TSLA["Adj Close"], color="y")
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title("TSLA")

ax[1,1].plot(FaceBook.index, FaceBook["Adj Close"], color="g")
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title("FaceBook")
```
