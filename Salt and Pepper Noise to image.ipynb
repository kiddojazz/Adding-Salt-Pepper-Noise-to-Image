```python
#Importing The necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
from PIL import Image
```

```python
def main():
    img = cv2.imread('IMG_2641.jpg', 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    rows, columns, channels = img.shape #Needed for calculating the dimension of the image
    p = 0.05 # 5% probability of noise
    output = np.zeros(img.shape, np.uint8)
    
    #We have to iterate through the image
    for i in range(rows):
        for j in range(columns):
            r = random.random()
            if r < p/2:
                #Pepper Sprinkled
                output[i][j] = [0, 0, 0] # black
            elif r < p:
                #salt sprinkled
                output[i][j] = [255, 255, 255]
            else:
                output[i][j] = img[i][j]
    
    plt.imshow(output)
    plt.savefig("Dotun_Salt/s&p3299.png")
    #cv2.imshow("Noisee",output)
    #cv2.waitKey(0)
    #cv2.imwrite("noisee.jpg",output)
#Before we can introduce salt and pepper noise we need to convert some region to complete white on the image
#Remember white is 255,255,255, while black is 0,0,0
    plt.title('Image with Salt and Pepper Noise')

    plt.show()

if __name__ == "__main__":

    main()
```
